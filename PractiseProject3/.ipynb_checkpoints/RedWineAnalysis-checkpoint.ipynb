### Introduction <a name='introduction'></a> 

<p>
    Red wine is a type of wine made from dark-colored grape varieties. The actual color of the wine can range from intense violet, typical of young wines, through to brick red for mature wines and brown for older red wines. The juice from most purple grapes is greenish-white, the red color coming from anthocyan pigments (also called anthocyanins) present in the skin of the grape; exceptions are the relatively uncommon teinturier varieties, which produce a red-colored juice. Much of the red-wine production process therefore involves extraction of color and flavor components from the grape skin. It is a delicacy around the world.
</p>

```python
#Importing all necessary packages
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from scipy import stats


from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier

from imblearn.over_sampling import RandomOverSampler

from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
from sklearn.metrics import auc, roc_curve, roc_auc_score
from sklearn.metrics import plot_roc_curve

from sklearn.model_selection import GridSearchCV

import joblib

import warnings
warnings.filterwarnings('ignore')
```

### Data and EDA <a name='data'></a>

<p>
    The data contains various physiochemical data as input variables while sensory data as output variable (column 'quality'). While the output variable contains discreet numbers ranging from 3 to 10, through some feature engineering, we will convert the output variable into a binary class with 1 for all wine with 'quality' above 7 and 0 for those below 7.

```python
wineDf = pd.read_csv('winequality-red.csv')
wineDf.head()
```

Except the columns 'quality', 'citric acid', 'density' and 'ph', all the columns have a lot of skewness which needs to be rectified. It is also worth noting that all the columns with skewness are positively skewed.

```python
plt.figure(figsize=(10,20))
for index in range(len(wineDf.columns)):
    plt.subplot(5,3, index+1)
    plt.boxplot(wineDf[wineDf.columns[index]])
    plt.xlabel(wineDf.columns[index])
    plt.tight_layout()
```

These boxplots above further confirm our suspecion about the presence of outliers. We will remove the outliers before processing further.

```python
#Removing outliers
z = np.abs(stats.zscore(wineDf))
threshold = 3
wineDf_new = wineDf[(z<3).all(axis=1)]
```

Now, with the code below, we will add a new column with the value 0 for all rows with 'quality' below 7 and 1 for 'quality' above 7.

```python
wineDf_new['quality_label'] = np.where(wineDf_new['quality'] >= 7, 1, 0)
wineDf_new.quality_label.value_counts()
```

We see that the 'quality_label' is imbalanced. This will be dealt with later in the project.

```python
wineDf_new.groupby('quality_label')['quality'].median().plot()
plt.show()
```

This diagram shows that the median for those with 0 in 'quality_label' is 5 and for those with 1 in 'quality_label' is 7. 

```python
figure = plt.subplots(5,3, figsize = (20,20))

for index in range(0, 12):
    plt.subplot(4,3,index+1)
    sns.scatterplot(wineDf[wineDf_new.columns[index]], wineDf_new['quality'], hue=wineDf_new['quality_label'])
    plt.tight_layout()
plt.show()
```

The above graphs reveal that the spread of the physiochemical values across the independant columns is more or less even irrespective of the wine have the 'quality_label' of 0 or not.

```python
plt.figure(figsize=(10,10))
sns.heatmap(wineDf_new.corr(), vmax=1, vmin=-1, annot=True)
plt.show()
```

The heatmap above confirms our observation from the scatter plots in the previous cell. All the independant columns are weakly correlated to the dependant column as evident from the above diagram.

### Analysis <a name='analysis'></a>

```python
#Seperating the independant and depandent columns
x = wineDf_new.drop('quality_label', axis = 1)
y = wineDf_new.quality_label
```

```python
#Removing the skewness of all the continuous columns and re-checking their skewness afterwards
from sklearn.preprocessing import power_transform

xt = power_transform(x)

wineDf_x_transformed = pd.DataFrame(xt, columns= x.columns)
wineDf_x_transformed.skew()
```

```python
#Scaling the independent features using MinMax Scaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
wineDf_x_transformed = scaler.fit_transform(wineDf_x_transformed)
```

```python
wineDf_x_final = pd.DataFrame(wineDf_x_transformed, columns=x.columns)
wineDf_x_final.head()
```

```python
#Finding the best random state
maxAcc = 0
maxRs = 0
for randState in range(0,200):
    x_train, x_test, y_train, y_test = train_test_split(wineDf_x_final, y, test_size=0.30, random_state=randState)
    lr = LogisticRegression()
    lr.fit(x_train, y_train)
    predlr = lr.predict(x_test)
    acc = accuracy_score(y_test, predlr)
    if acc>maxAcc:
        maxAcc=acc
        maxRs=randState
print(f"Best accuracy is {maxAcc*100}% on Random state {randState}")
```

```python
x_train, x_test, y_train, y_test = train_test_split(wineDf_x_final, y, test_size=0.30, random_state=199)
```

```python
#Oversampling tp balance the dataset

ros = RandomOverSampler(random_state = 42)
x_resampled, y_resampled = ros.fit_resample(x_train, y_train)

print(f'Before Resampling\n{y_train.value_counts()}\nAfter Resampling:\n{y_resampled.value_counts()}')
```

#### Trying different classification algorithms
1. KNeighborClassifier

```python
knn = KNeighborsClassifier()
knn.fit(x_resampled, y_resampled)

predKnn = knn.predict(x_test)
predKnn_proba = knn.predict_proba(x_test)

print(f'Confusion Matrix:\n{confusion_matrix(y_test, predKnn)}\n')
print(f'Classification Report:\n{classification_report(y_test, predKnn)}\n')
print(f'F1 Score: \n{f1_score(y_test, predKnn)}\n')
print(f'The ROC-AUC Score is : {roc_auc_score(y_test, predKnn_proba[:,1])}')
```

2. Random Forest Classifier

```python
rfc = RandomForestClassifier()
rfc.fit(x_resampled, y_resampled)
predRfc = rfc.predict(x_test)
predRfc_proba = rfc.predict_proba(x_test)

print(f'Confusion Matrix:\n{confusion_matrix(y_test, predRfc)}\n')
print(f'Classification Report:\n{classification_report(y_test, predRfc)}\n')
print(f'F1 Score: \n{f1_score(y_test, predRfc)}\n')
print(f'The ROC-AUC Score is : {roc_auc_score(y_test, predRfc_proba[:,1])}')
```

3. Support Vector Classifier

```python
svc = SVC(probability=True)
svc.fit(x_resampled, y_resampled)
predSvc = svc.predict(x_test)
predSvc_proba = svc.predict_proba(x_test)

print(f'Confusion Matrix:\n{confusion_matrix(y_test, predSvc)}\n')
print(f'Classification Report:\n{classification_report(y_test, predSvc)}\n')
print(f'F1 Score: \n{f1_score(y_test, predSvc)}\n')
print(f'The ROC-AUC Score is : {roc_auc_score(y_test, predSvc_proba[:,1])}')
```

4. AdaBoost Classifier

```python
ab = AdaBoostClassifier()
ab.fit(x_resampled, y_resampled)

predAb = ab.predict(x_test)
predAb_proba = ab.predict_proba(x_test)

print(f'Confusion Matrix:\n{confusion_matrix(y_test, predAb)}\n')
print(f'Classification Report:\n{classification_report(y_test, predAb)}\n')
print(f'F1 Score: \n{f1_score(y_test, predAb)}\n')
print(f'The ROC-AUC Score is : {roc_auc_score(y_test, predAb_proba[:,1])}')
```

```python
ax = plt.gca()
knn_disp = plot_roc_curve(knn, x_test, y_test, ax=ax, alpha=0.8)
rfc_disp = plot_roc_curve(rfc, x_test, y_test, ax=ax, alpha=0.8)
svc_disp = plot_roc_curve(svc, x_test, y_test, ax=ax, alpha=0.8)
ab_disp = plot_roc_curve(ab, x_test, y_test, ax=ax, alpha=0.8)
plt.show()
```

```python
cvKnn = cross_val_score(knn, x_resampled, y_resampled, n_jobs=-1, cv=5, scoring='roc_auc')
print(f'Cross val score is: {cvKnn.mean()}')
print(f'Difference between roc-auc score and crossval roc-auc score: {(roc_auc_score(y_test, predKnn_proba[:,1])-cvKnn.mean())}')
```

```python
cvRfc = cross_val_score(rfc, x_resampled, y_resampled, n_jobs=-1, cv=5, scoring='roc_auc')
print(f'Cross val score is: {cvRfc.mean()}')
print(f'Difference between roc-auc score and crossval roc-auc score: {(roc_auc_score(y_test, predRfc_proba[:,1])-cvRfc.mean())}')
```

```python
cvSvc = cross_val_score(svc, x_resampled, y_resampled, n_jobs=-1, cv=5, scoring='roc_auc')
print(f'Cross val score is: {cvSvc.mean()}')
print(f'Difference between roc-auc score and crossval roc-auc score: {(roc_auc_score(y_test, predSvc_proba[:,1])-cvSvc.mean())}')
```

```python
cvAb = cross_val_score(ab, x_resampled, y_resampled, n_jobs=-1, cv=5, scoring='roc_auc')
print(f'Cross val score is: {cvAb.mean()}')
print(f'Difference between roc-auc score and crossval roc-auc score: {(roc_auc_score(y_test, predAb_proba[:,1])-cvAb.mean())}')
```

<p>
    From the table above, we can see that the difference in ROC-AUC score and cross-validation RUC-AUC score is least in case of KNeighborClassifier. Therefore, out of the three models tried in this analysis, this is the best model for the given dataset. However, we can still try to increase the accuracy of this model by hyper parameter tuning. We shall do this in the next few cells.
</p>

#### Hyper Parameter Tuning

```python
parameters = {'n_neighbors': [10,20,40,80,100,150,200],
             'weights': ['distance'],
             'algorithm': ['ball_tree', 'kd_tree', 'brute'],
             'p': [1]}
```

```python
gs = GridSearchCV(knn, param_grid = parameters, scoring= 'roc_auc')
gs.fit(x_resampled, y_resampled)
gs.best_params_
```

Using the best parameter conbination found above to make our final model using KNeighborsClassifier.

```python
knn = KNeighborsClassifier(algorithm = 'ball_tree', n_neighbors = 20, p = 1, weights = 'distance')
knn.fit(x_resampled, y_resampled)
knnPred = knn.predict_proba(x_test)
print(f'The final RUC-AUC score is {roc_auc_score(y_test, knnPred[:,1])}')
```

Finding the threshold levels

```python
fpr, tpr, thresholds = roc_curve(y_test, knnPred[:,1])
thresholds
```

```python
accuracy_ls = []
for thres in thresholds:
    y_pred = np.where(knnPred[:,1]>thres,1,0)
    accuracy_ls.append(accuracy_score(y_test, y_pred, normalize=True))
    
accuracy_ls = pd.concat([pd.Series(thresholds), pd.Series(accuracy_ls)],
                        axis=1)
accuracy_ls.columns = ['thresholds', 'accuracy']
accuracy_ls.sort_values(by='accuracy', ascending=False, inplace=True)
accuracy_ls
```

The above threshold can be used to optimise the model to balance the true-positive rate and the false-positive rate depending on the requirement of the model.
