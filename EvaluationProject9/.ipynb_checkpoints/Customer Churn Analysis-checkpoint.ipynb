## Introduction <a name = "1"></a>

For all firms, while it is important to acquire new customers, it is also equally crucial that they retain their exixting customer. In fact, in most cases, it is more expensive to acquire new customer then retaining existing customer as acquiring new customers involve the cost of marketing, etc. In order to retain customers, a company needs the ability to predict customer churn so that those likely to discontinue their business with the firm can be identifed in advance and can be provided enchanced service in order to retain them.

This analysis aims to achieve just that by building a model to predict customer churn.

## Analysis <a name = '4'></a>

```python
numericCol.remove('SeniorCitizen')
categoricalCol.append('SeniorCitizen')
```

```python
x = customerDf.drop('Churn', axis = 1)
y = customerDf['Churn']
```

```python
#Applying Standard Scaler to numeric columns
colTrans = make_column_transformer((StandardScaler(), numericCol),(OneHotEncoder(), categoricalCol), remainder='passthrough')
x = colTrans.fit_transform(x)
```

```python
y = np.where(y == 'Yes', 1, 0)
```

```python
maxAcc = 0
maxRs = 0
for randState in range(0,200):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=randState)
    lr = LogisticRegression()
    lr.fit(x_train, y_train)
    predlr =  lr.predict(x_test)
    acc = lr.score(x_test, y_test)
    if acc>maxAcc:
        maxAcc=acc
        maxRs=randState
print(f"Best accuracy is {maxAcc*100}% on Random state {randState}")
```

```python
#splitting into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=199)
```

```python
#Oversampling to balance the dataset
ros = RandomOverSampler(sampling_strategy = 0.7,random_state = 42)
xResampled, yResampled = ros.fit_resample(x_train, y_train)
```

```python
#Training various classification models and checking their accuracies
lr = LogisticRegression()
rfc = RandomForestClassifier()
dtc = DecisionTreeClassifier() 
svc = SVC() 
knc = KNeighborsClassifier()
adc = AdaBoostClassifier()

models = [lr, rfc, dtc, svc, knc, adc]

for model in models:
    model.fit(xResampled, yResampled)
    pred = model.predict(x_test)
    pred_train = model.predict(x_train)
    print(f'\t{model}\nConfusion Matrix:\n{confusion_matrix(y_test, pred)}\n')
    print(f'Classification Report:\n{classification_report(y_test, pred)}\n')
    print(f'F1 Score: \n{f1_score(y_test, pred)}\n')
    print(f'Accuracy Score: {accuracy_score(y_test, pred)}\n')
    print(f'Train and test score difference: {accuracy_score(y_train, pred_train) - accuracy_score(y_test, pred)}')
```

SVC has the least test and training score difference. Hence, SVC is better able to generalise the dataset to make better predictions.

##### Cross Validation

```python
svcCV = cross_val_score(SVC(), xResampled, yResampled, cv = 5)
print(f'Cross val score is: {svcCV.mean()}')
```

## Hyper Parameter Tuning <a name = "5"></a>

```python
parameterSVR = {'kernel': ['linear', 'poly', 'rbf'],
               'gamma': ['scale', 'auto'],
               'C': [1, 0.1, 0.001, 0.0001]}
```

```python
svcGS = GridSearchCV(SVC(), cv = 5, param_grid = parameterSVR, n_jobs = -1)
```

```python
svcGS.fit(xResampled, yResampled)
```

```python
svcGS.best_params_
```

```python
svc = SVC(C = 1, gamma = 'scale', kernel = 'linear')
```

```python
svc.fit(xResampled, yResampled)
```

```python
pred = svc.predict(x_test)
print(f'Accuracy Score: {accuracy_score(y_test, pred)}\n')
```
