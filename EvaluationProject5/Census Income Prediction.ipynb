```python
def chi2IndependenceTest(categoricalColName):
    
    crossTab = pd.crosstab(incomeDf[categoricalColName], incomeDf['Income'], margins=True, margins_name='Total')
    observedVal = crossTab.values
    expectedVal = stat.chi2_contingency(crossTab)[3]
    rowNum = (crossTab.shape[0] - 1)
    columnNum = (crossTab.shape[1] - 1)
    doF=(rowNum-1)*(columnNum-1)
    chiSquare=sum([(o-e)**2./e for o,e in zip(observedVal, expectedVal)])
    chiSquareStatistic=chiSquare[0]+chiSquare[1]
    pValue=1-stat.chi2.cdf(x=chiSquareStatistic,df=doF)
    
    return doF, pValue
```

```python
alpha = 0.05
sigCatCols = [] #List of significant categorical columns

for cols in categoricalCols:
    degreesOfFreedom, pValue = chi2IndependenceTest(cols)
    if pValue<=alpha:
        sigCatCols.append(cols)
        print(f'Reject H0, There is a relationship between {cols} and Income')
    else:
        print(f'Retain H0,There is no relationship between {cols} and Income')
```

All the categorical columns have a statistically significant relation with the target column.

## Preprocessing and Analysis <a name = '4'></a>

```python
incomeDf[categoricalCols]['Workclass'].unique()
```

We will treat the Workclass column as ordinal categorical column. This is because, usually those belonging to Private sector earn more followed by those in government and lastly those with 'Without-pay' and 'Never-worked' will earn the least. So the Private labels will get encoded a higher number so that it recieves more importance in the model followed by govenment job holders and so on.

```python
workclassDict = {' Private': 4,
                ' Self-emp-not-inc': 2,
                ' Local-gov': 3,
                ' State-gov': 3,
                ' Self-emp-inc': 2,
                ' Federal-gov': 3,
                ' Without-pay': 1,
                ' Never-worked': 1}
```

```python
incomeDf['Workclass'] = incomeDf['Workclass'].map(workclassDict)
```

```python
#Dropping the education column because the column 'Education_num' is the ordinally encoded version of this column
#and will be used in the model
incomeDf.drop('Education', axis = 1, inplace = True)
```

```python
#OneHotEncoding the rest of the categorical columns
incomeDf = pd.get_dummies(incomeDf, columns = ['Sex', 'Marital_status', 'Occupation', 'Relationship', 'Race'])
```

```python
incomeDf['Native_country'].unique()
```

Since the 'Native_country' column contains a lot of countries, onehotencoding them will increase the dataset dimensions immensely. Therefore we will categorise and ordinally encode these countries according to the development stage they are in currently. The underdeveloped countries will have the least number and the developed countries will have the highest number.

```python
# 3 = developed, 2 = developing, 1 = underdeveloped
countryDict = {
    ' United-States':3,
    ' Cuba': 2,
    ' Jamaica': 3,
    ' India': 2,
    ' Mexico': 2,
    ' South':2,
    ' Puerto-Rico':3,
    ' Honduras':2,
    ' England':3,
    ' Canada':3,
    ' Germany':3,
    ' Iran':2,
    ' Philippines':2,
    ' Italy':3,
    ' Poland':3,
    ' Columbia':1,
    ' Cambodia':2,
    ' Thailand':2,
    ' Ecuador':2,
    ' Laos':1,
    ' Taiwan':3,
    ' Haiti':2,
    ' Portugal':3,
    ' Dominican-Republic':2,
    ' El-Salvador':2,
    ' France':3,
    ' Guatemala':2,
    ' China':2,
    ' Japan':3,
    ' Yugoslavia':1,
    ' Peru':2,
    ' Outlying-US(Guam-USVI-etc)':2,
    ' Scotland':3,
    ' Trinadad&Tobago':1,
    ' Greece':3,
    ' Nicaragua':2,
    ' Vietnam':2,
    ' Hong':3,
    ' Ireland':3,
    ' Hungary':3,
    ' Holand-Netherlands':3
}
```

```python
incomeDf['Native_country'] = incomeDf['Native_country'].map(countryDict)
```

```python
#Making a function to detect outliers using the IQR method and then replacing them with the upper and lower bridge values
def outlierRemoverS(col, data = incomeDf):
    '''Note: This function is for skewed data only'''
    IQR = data[col].quantile(0.75) - data[col].quantile(0.25)
    lowerBridge = data[col].quantile(0.25) - (1.5*IQR)
    upperBridge = data[col].quantile(0.75) + (1.5*IQR)
    if lowerBridge >=0:
            data.loc[data[col]>=upperBridge, col] = upperBridge
            data.loc[data[col]<=lowerBridge, col] = lowerBridge
    elif lowerBridge < 0:
            data.loc[data[col]>=upperBridge, col] = upperBridge
            data.loc[data[col]<=0, col] = data[col].quantile(0.5)
```

```python
#Removing outliers
outlierRemoverS('Hours_per_week')
```

```python
#Seperating the X and Y
x = incomeDf.drop('Income', axis = 1)
y = np.where(incomeDf['Income'] == ' <=50K', 1, 2)
```

```python
#Applying Standard Scaler to numeric columns
colTrans = make_column_transformer((StandardScaler(), ['Age', 'Fnlwgt','Hours_per_week', 'Capital_gain', 'Capital_loss']), remainder='passthrough')
x = colTrans.fit_transform(x)
```

```python
#Finding the best random state
maxAcc = 0
maxRs = 0
for randState in range(0,200):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=randState)
    lr = LogisticRegression()
    lr.fit(x_train, y_train)
    predlr =  lr.predict(x_test)
    acc = lr.score(x_test, y_test)
    if acc>maxAcc:
        maxAcc=acc
        maxRs=randState
print(f"Best accuracy is {maxAcc*100}% on Random state {randState}")
```

```python
#splitting into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=199)
```

```python
#Oversampling to balance the dataset
ros = RandomOverSampler(sampling_strategy = 0.7,random_state = 42)
xResampled, yResampled = ros.fit_resample(x_train, y_train)
```

```python
#Training various classification models and checking their accuracies
lr = LogisticRegression()
rfc = RandomForestClassifier()
dtc = DecisionTreeClassifier() 
svc = SVC() 
knc = KNeighborsClassifier()
adc = AdaBoostClassifier()

models = [lr, rfc, dtc, svc, knc, adc]

for model in models:
    model.fit(xResampled, yResampled)
    pred = model.predict(x_test)
    print(f'\t{model}\nConfusion Matrix:\n{confusion_matrix(y_test, pred)}\n')
    print(f'Classification Report:\n{classification_report(y_test, pred)}\n')
    print(f'F1 Score: \n{f1_score(y_test, pred)}\n')
    print(f'Accuracy Score: {accuracy_score(y_test, pred)}\n')
```

#### Checking Cross Validation Scores

We will chose RandomForestClassifier as our final model since it has the highest accuracy and also the least difference in score after cross validation.
