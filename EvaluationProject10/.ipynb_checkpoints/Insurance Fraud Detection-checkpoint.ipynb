## Introduction <a name = '1'></a>

Insurance is a very secure and reliable way to ensure any unpredictable and unplesant future event can be tackled wihtout causing huge financial loss. It is a way of secure one's future. When an individual purchases an insurance policy, the insurance company guarantees to pay out the individual a agreed sum of money for whatever the insurance was for. In case of car insurance, if a vehicle is met with an accident or needs repairing, the insurance company will bear the cost for the repairs. However, this leaves an avenue to bilk the insurance companies of their money by reporting false repairs and appropriating money from them. This is called Insurance Fraud. 

This model aims to identify insurance fraud using the data provided.

We see that some rows have '?' as labels which indicates missing values. We will first impute these missng values and then carry forwards.

```python
#Making a function to inpute categorical columns
def missingImputerCat(colName):
    try:
        insuranceDf[f'{colName}_missing'] = np.where(insuranceDf[colName] == '?', 1, 0)
        insuranceDf[colName] = np.where(insuranceDf[colName] == '?', insuranceDf[colName].mode(), insuranceDf[colName])
    except:
        insuranceDf[f'{colName}_missing'] = np.where(insuranceDf[colName] == '?', 1, 0)
        insuranceDf[colName] = np.where(insuranceDf[colName] == '?', insuranceDf[colName].mode()[1], insuranceDf[colName])
```

```python
#Applying the function above to all columns with '?' in them
for index in categoricalCols:
    if '?' in insuranceDf[index].unique():
        missingImputerCat(index)
```

```python
#Defining a function to do all of the above steps and tell us if the target column has any significant relation with other categorical column
def chi2IndependenceTest(categoricalColName):
    
    crossTab = pd.crosstab(insuranceDf[categoricalColName], insuranceDf['fraud_reported'], margins=True, margins_name='Total')
    observedVal = crossTab.values
    expectedVal = stats.chi2_contingency(crossTab)[3]
    rowNum = (crossTab.shape[0] - 1)
    columnNum = (crossTab.shape[1] - 1)
    doF=(rowNum-1)*(columnNum-1)
    chiSquare=sum([(o-e)**2./e for o,e in zip(observedVal, expectedVal)])
    chiSquareStatistic=chiSquare[0]+chiSquare[1]
    pValue=1-stats.chi2.cdf(x=chiSquareStatistic,df=doF)
    
    return doF, pValue
```

```python
alpha = 0.05
sigCatCols = [] #List of significant categorical columns

for cols in categoricalCols:
    degreesOfFreedom, pValue = chi2IndependenceTest(cols)
    if pValue<=alpha:
        sigCatCols.append(cols)
        print(f'Reject H0, There is a relationship between {cols} and Loan_Status')
    else:
        print(f'Accept H0,There is no relationship between {cols} and Loan_Status')
```

```python
sigCatCols
```

The above columns have a statistically significant relation with the target variable

```python
#Making a function to detect outliers using the IQR method and then replacing them with the upper and lower bridge values
def outlierRemoverS(data, col):
    '''Note: This function is for skewed data only'''
    IQR = data[col].quantile(0.75) - data[col].quantile(0.25)
    lowerBridge = data[col].quantile(0.25) - (1.5*IQR)
    upperBridge = data[col].quantile(0.75) + (1.5*IQR)
    if lowerBridge >=0:
            data.loc[data[col]>=upperBridge, col] = upperBridge
            data.loc[data[col]<=lowerBridge, col] = lowerBridge
    elif lowerBridge < 0:
            data.loc[data[col]>=upperBridge, col] = upperBridge
            data.loc[data[col]<=0, col] = data[col].quantile(0.5)
```

```python
for i in ['age', 'umbrella_limit', 'property_claim']:
    outlierRemoverS(insuranceDf, i)
```

## Analysis <a name = '4'></a>

```python
x = insuranceDf.drop('fraud_reported', axis = 1)
y = insuranceDf['fraud_reported']
```

```python
#making a column transformer to perform standard scaling on numeric columns, one hot encoding on categorical columns and label encoder in target column
colTrans = make_column_transformer((StandardScaler(), numericCols),(OneHotEncoder(), sigCatCols), remainder='drop') 
x = colTrans.fit_transform(x)
```

```python
y = np.where(y == 'Y', 1, 0)
```

```python
#Finding the best random state
maxAcc = 0
maxRs = 0
for randState in range(0,200):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=randState)
    lr = LogisticRegression()
    lr.fit(x_train, y_train)
    predlr = lr.predict(x_test)
    acc = lr.score(x_test, y_test)
    if acc>maxAcc:
        maxAcc=acc
        maxRs=randState
print(f"Best accuracy is {maxAcc*100}% on Random state {randState}")
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=199)
```

```python
#Oversampling to balance the dataset

ros = RandomOverSampler(random_state = 42)
xResampled, yResampled = ros.fit_resample(x_train, y_train)
```

```python
#Training various classification models and checking their accuracies
lr = LogisticRegression()
rfc = RandomForestClassifier()
dtc = DecisionTreeClassifier() 
svc = SVC() 
adc = AdaBoostClassifier()

models = [lr, rfc, dtc, svc, adc]

for model in models:
    model.fit(xResampled, yResampled)
    pred = model.predict(x_test)
    print(f'\t{model}\nConfusion Matrix:\n{confusion_matrix(y_test, pred)}\n')
    print(f'Classification Report:\n{classification_report(y_test, pred)}\n')
    print(f'F1 Score: \n{f1_score(y_test, pred)}\n')
    print(f'Accuracy Score: {accuracy_score(y_test, pred)}\n')
```

```python
lrCV = cross_val_score(lr , xResampled, yResampled, cv = 5)
print(f'Cross val score is: {lrCV.mean()}')
print(f'Accuracy and Cross Validation Score difference: {0.8466666666666667 - lrCV.mean()}')
```

```python
rfcCV = cross_val_score(rfc , xResampled, yResampled, cv = 5)
print(f'Cross val score is: {rfcCV.mean()}')
print(f'Accuracy and Cross Validation Score difference: {0.81 - rfcCV.mean()}')
```

```python
dtcCV = cross_val_score(dtc , xResampled, yResampled, cv = 5)
print(f'Cross val score is: {dtcCV.mean()}')
print(f'Accuracy and Cross Validation Score difference: {0.7833333333333333 - dtcCV.mean()}')
```

```python
svcCV = cross_val_score(svc , xResampled, yResampled, cv = 5)
print(f'Cross val score is: {svcCV.mean()}')
print(f'Accuracy and Cross Validation Score difference: {0.81 - svcCV.mean()}')
```

```python
adcCV = cross_val_score(adc , xResampled, yResampled, cv = 5)
print(f'Cross val score is: {adcCV.mean()}')
print(f'Accuracy and Cross Validation Score difference: {0.82 - adcCV.mean()}')
```

Since RandomForestClassifier has the highest accuracy after cross validation and also the difference of -0.11 (approx) means that the cross validation score has increase by 0.11. By the same logic, DecisionTreeClassifier also seems like a good model. Therefore, we will perform hyperparameter tuning on both and then decide the final model.
