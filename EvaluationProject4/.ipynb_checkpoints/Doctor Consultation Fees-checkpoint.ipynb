## Introduction: The Problem

Our health is the most important aspect of our daily lives. Phrases like 'Health is wealth' is often quoted. Therefore, apart from doctor visits while sick, in order to maintain a good health, it is often recommended to visit a doctor at regular intervals even when someone is not ill just for health checks. HOwever, doctor visits can throw some unexpected surprises in the form of high consultation fees.

In this project we will try to predict the doctor's consultation fees using machine learning.

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab 


from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import mean_squared_error, r2_score, mean_squared_log_error

import joblib

import warnings
warnings.filterwarnings('ignore')
```

## Exploratory Data Analysis <a name = '3'></a>

Univariate Analysis

```python
df.isnull().sum()
```

The dataset has 3302 missing values in Rating, 25 in Place and 2620 in Miscellaneous_Info. We will deal with these missing values first before moving further.

```python
#Making list of missing values
missingCols = [i for i in df.columns if df[i].isnull().sum() >= 1]
missingCols
```

Now we will define a function that will take a column as input and the make a new column containing the word 'Missing' for the rows where the input column had missing values and will contain in original values for the rows that did not have a missing value. This was, we can preserve any relation the target column might have with the value column. It will then drop the orginal column.

```python
def catMissingImputer(colName):
    df[colName+"_Imputed"]=np.where(df[colName].isnull(),"Missing",df[colName])
    df.drop(colName, axis = 1, inplace = True)
```

```python
#Applying the function to all missing columns
for column in missingCols:
    catMissingImputer(column)
```

```python
df.head()
```

Now, we shall preform some preprocessing in order to make the columns more suited for data modelling.

```python
#Keeping on the numeric years and discarding the 'years experience'
df.Experience = df['Experience'].apply(lambda x: x.split()[0])
```

```python
#Keeping on the numeric rating and discarding the % sign
df.Rating_Imputed = df['Rating_Imputed'].apply(lambda x: x.split('%')[0])
```

```python
#Making a new column to store the locality of the doctor clinic
df.Place_locality = df['Place_Imputed'].apply(lambda x: x.split(',')[0])
```

```python
df.Place_Imputed = np.where(df.Place_Imputed == 'Missing', 'Missing, Missing', df.Place_Imputed)
```

Running this code will throw error as there is some value which has only one word and no ','. Therefore there is no element in the index 1 and when we try to extract the 1 index after the split using ',', it will throw IndexError. The following lines of code identifies that row/rows and then replaces it with the mode of the column.

```python
#df.Place_city = df['Place_Imputed'].apply(lambda x: x.split(',')[1])
```

```python
for i in df.Place_Imputed.unique():
    if ',' not in i:
        print(i)
```

Therefore the value which is causing our error is 'e'. Using the code below we can find the row where it is present.

```python
df[df['Place_Imputed'] == 'e']
```

```python
#Finding the mode for all those with BDS Wualification and Profile Dentist to replace the e above
print(df[(df['Qualification'] == 'BDS') & (df['Profile'] == 'Dentist')]['Place_Imputed'].mode())

df.Place_Imputed = np.where(df.Place_Imputed == 'e', 'R.S. Puram, Coimbatore', df.Place_Imputed)
```

```python
#Making a new column to store the city the clinic is located
df.Place_city = df['Place_Imputed'].apply(lambda x: x.split(',')[1])
```

The specilization of the doctor is captured in the profile column so in the qualification column we can just keep the first word before the ','. This is done using the code below.

```python
df['Qualification_new'] = df['Qualification'].apply(lambda x: x.split(',')[0])
```

Since we have extracted the necessary information from the Qualification column, we will now drop it. We will also drop the Miscellaneous_Info_Imputed' column because the information in it does not have consistency and will require natural language processing of some sort to gain meaningful insigths.

```python
df.drop(['Qualification', 'Miscellaneous_Info_Imputed'], axis = 1, inplace=True)
```

```python
df.dtypes
```

The datatypes of the colum Experience and Rating_Imputed can be changed to int64 as they only contain integers now. But before that, we have to handle all the rows in the Rating_Imputed column which has 'Missing' as entry. In order to do that, we will make a new column where if there was a missing value in the Rating_Imputed column, then the new column will have 1 or else 0. In the original Rating_Imputed column, we will replace the 'Missing' with 0 and then convert the datatypes of the aformentioned columns.

```python
df['Rating_Missing'] = np.where(df.Rating_Imputed == 'Missing', 1, 0)
df.Rating_Imputed = np.where(df.Rating_Imputed == 'Missing', 0, df.Rating_Imputed)
```

```python
df = df.astype({'Experience': 'int64', 'Rating_Imputed': 'int64'})
```

```python
#function to make QQ plot so that we can judge the Gaussion distribution of the numeric columns
def dataPlot(df,feature):
    plt.figure(figsize=(20,10))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

## Preprocessing and Analysis <a name = '4'></a>

```python
df.skew()
```

```python
#Removing skewness and checking QQ plot again
df['Experience'] = (df['Experience'])**(1/2)
```

```python
df.skew()
```

```python
dataPlot(df, 'Experience')
```

```python
#Splitting the target and feature columns
x = df.drop('Fees', axis = 1)
y = df.Fees
```

```python
#Making a list of top 10 qualifications
topQualification10=x.Qualification_new.value_counts().sort_values(ascending=False).head(10).index
topQualification10=list(topQualification10)
```

```python
#One hot encoding the top 10 qualifications
for categories in topQualification10:
    x[categories]=np.where(x['Qualification_new']==categories,1,0)
```

```python
#Dropping Qualification_new as it wont be needed anymore 
x.drop('Qualification_new', axis = 1, inplace=True)
```

```python
x
```

```python
#One hot encoding the top 10 clinic location in the Place_Imputed column
x.Place_Imputed.value_counts()
```

```python
topPlace10=df.Place_Imputed.value_counts().sort_values(ascending=False).head(10).index
topPlace10=list(topPlace10)
```

```python
for categories in topPlace10:
    x[categories]=np.where(x['Place_Imputed']==categories,1,0)
```

```python
x.drop('Place_Imputed', axis = 1, inplace=True)
```

```python
#One hot encoding the PRofile column
x = pd.get_dummies(x, columns=['Profile'])
```

```python
#making a column transformer to perform standard scaling on numeric columns
colTrans = make_column_transformer((StandardScaler(), ['Experience', 'Rating_Imputed']), remainder='passthrough') 
x = colTrans.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
#Fitting various regression models to the train data and printing their perfomance metrics one-by-one
lr = LinearRegression()
rfr = RandomForestRegressor()
dtr = DecisionTreeRegressor()
svr = SVR()
knr = KNeighborsRegressor()
mlp = MLPRegressor()
br = BaggingRegressor()
gb = GradientBoostingRegressor()

models = [lr, rfr, dtr, svr, knr, mlp, br, gb]

for model in models:
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    df1 = pd.DataFrame()
    print(f'{model}\n\tAccuracy: {model.score(x_test, y_test)}\n\tMSLE: {mean_squared_log_error(y_test, pred)}')
```

For this analysis, we will use Mean Square Log Error it will penalize large errors than small ones. R2 score will not give a good measure as seen from the scores above. We will choose DecisionTreeRegressor as it has the highest MSLE and perform hyper parameter tuning.

#### Cross Validation

```python
svrCV = cross_val_score(svr, x_train, y_train, cv = k_fold, scoring = 'neg_mean_squared_log_error')
print(f'CV score for {svr}: {-svrCV.mean()}')
```

```python
knrCV = cross_val_score(knr, x_train, y_train, cv = k_fold, scoring = 'neg_mean_squared_log_error')
print(f'CV score for {knr}: {-knrCV.mean()}')
```

```python
mlpCV = cross_val_score(mlp, x_train, y_train, cv = k_fold, scoring = 'neg_mean_squared_log_error')
print(f'CV score for {mlp}: {-mlpCV.mean()}')
```

```python
brCV = cross_val_score(br, x_train, y_train, cv = k_fold, scoring = 'neg_mean_squared_log_error')
print(f'CV score for {br}: {-brCV.mean()}')
```

```python
gbCV = cross_val_score(gb, x_train, y_train, cv = k_fold, scoring = 'neg_mean_squared_log_error')
print(f'CV score for {gb}: {-gbCV.mean()}')
```

## Hyper Parameter Tuning <a name = '5'></a>

```python
parameters = {'criterion': ['mse', 'mae', 'poisson', 'friedman_mse'],
              'splitter': ['best', 'random'],
             'max_depth': [None, 5, 10, 15, 20, 50, 100, 150, 200, 500, 1000, 10000],
             'max_features': ['auto', 'sqrt', 'log2']}
```

```python
gs = GridSearchCV(dtr, param_grid = parameters, cv = 5, n_jobs = -1, scoring = 'neg_mean_squared_log_error')
```

```python
gs.fit(x_train, y_train)
```

```python
-gs.best_score_
```

```python
dtr = DecisionTreeRegressor(criterion = 'mae', max_depth = 5, max_features = 'auto', splitter = 'best')
dtr.fit(x_train, y_train)
pred = dtr.predict(x_test)
```

```python
print(f'MSLE: {mean_squared_log_error(y_test, pred)}')
```

```python
#Saving the model
joblib.dump(dtr, 'doctorFee.pkl')
```

## Final Predictions from Final_test.csv <a name = '6'></a>

```python
testdf = pd.read_excel('Final_Test.xlsx')

#Performing all preprocessing steps as done on the train file
testdf.isnull().sum()

#Making list of missing values
missingCols = [i for i in testdf.columns if testdf[i].isnull().sum() >= 1]
missingCols

def catMissingImputer(colName):
    testdf[colName+"_Imputed"]=np.where(testdf[colName].isnull(),"Missing",testdf[colName])
    testdf.drop(colName, axis = 1, inplace = True)

#Applying the function to all missing columns
for column in missingCols:
    catMissingImputer(column)

#Keeping on the numeric years and discarding the 'years experience'
testdf.Experience = testdf['Experience'].apply(lambda x: x.split()[0])

#Keeping on the numeric rating and discarding the % sign
testdf.Rating_Imputed = testdf['Rating_Imputed'].apply(lambda x: x.split('%')[0])

#Making a new column to store the locality of the doctor clinic
testdf.Place_locality = testdf['Place_Imputed'].apply(lambda x: x.split(',')[0])

testdf.Place_Imputed = np.where(testdf.Place_Imputed == 'Missing', 'Missing, Missing', testdf.Place_Imputed)

testdf.Place_city = testdf['Place_Imputed'].apply(lambda x: x.split(',')[1])

testdf['Qualification_new'] = testdf['Qualification'].apply(lambda x: x.split(',')[0])

testdf['Rating_Missing'] = np.where(testdf.Rating_Imputed == 'Missing', 1, 0)
testdf.Rating_Imputed = np.where(testdf.Rating_Imputed == 'Missing', 0, testdf.Rating_Imputed)

testdf = testdf.astype({'Experience': 'int64', 'Rating_Imputed': 'int64'})

testdf['Experience'] = (testdf['Experience'])**(1/2)

#Making a list of top 10 qualifications
topQualification10=testdf.Qualification_new.value_counts().sort_values(ascending=False).head(10).index
topQualification10=list(topQualification10)

#One hot encoding the top 10 qualifications
for categories in topQualification10:
    testdf[categories]=np.where(testdf['Qualification_new']==categories,1,0)

#Dropping Qualification_new as it wont be needed anymore 
testdf.drop(['Qualification','Qualification_new', 'Miscellaneous_Info_Imputed'], axis = 1, inplace=True)

topPlace10=testdf.Place_Imputed.value_counts().sort_values(ascending=False).head(10).index
topPlace10=list(topPlace10)

for categories in topPlace10:
    testdf[categories]=np.where(testdf['Place_Imputed']==categories,1,0)

testdf.drop('Place_Imputed', axis = 1, inplace=True)

#One hot encoding the PRofile column
testdf = pd.get_dummies(testdf, columns=['Profile'])

#making a column transformer to perform standard scaling on numeric columns
colTrans = make_column_transformer((StandardScaler(), ['Experience', 'Rating_Imputed']), remainder='passthrough') 
testdf = colTrans.fit_transform(testdf)
```

```python
model = joblib.load('doctorFee.pkl')
finalPred = model.predict(testdf)
```

```python
prediction = pd.DataFrame(finalPred, columns=['Fees'])
```
