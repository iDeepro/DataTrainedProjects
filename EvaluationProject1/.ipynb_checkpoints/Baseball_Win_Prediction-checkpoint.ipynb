## Introduction: The Problem <a name='intro'></a>

Predictive modelling has been used across all industries. The sports industry is not an exception. There are a variety of sports firms who use data science. Various media outlets and sports teams employ data scientists to predict game outcomes, study opponent team patterns, etc.

In this project, we will used data to predict how many wins a team will have based on previous data. Specifically, this project uses 2014 Major League Baseball seasons data in order to develop an algorithm that predicts the number of wins for a given team in the 2015 season.

```python
#importing necessary libraries
import numpy as np
import pandas as pd
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor

from sklearn.model_selection import GridSearchCV

from sklearn.metrics import mean_squared_error, r2_score

import warnings
warnings.filterwarnings('ignore')
```

## Exploratory Data Analysis <a name='eda'></a>

###### Univariate Analysis

```python
baseballDf.isnull().sum()
```

The data has no missing values.

```python
baseballDf.dtypes
```

All the columns are numeric columns and have the right datatypes. The column 'ERA', however, is of float64 datatype but since represents average values, it also has the right datatype.

```python
#Checking summary statistics
baseballDf.describe()
```

From the summary statistics, we see that for most columns the mean is more or less equal to the 50th Percentile value. Also the difference between the 75th Percentile and max value is not substantial. Therefore, it is more likely that there might not be any outliers. However, to be sure of it, we will use boxplot later in the section to check for outliers.

```python
#Forming a numeric column list and checking the distribution of each column
numericCols = list(baseballDf.drop('W', axis=1).columns)

plt.figure(figsize=(20,20))

for index in range(len(numericCols)):
    plt.subplot(4,4,index+1)
    sns.distplot(baseballDf[numericCols[index]])
    plt.tight_layout()
```

From the above distribution plots we see that most columns are following normal distribution. However, column R, H, CG, SHO, SV and E seems to have bimodel distribution. Furthermore, a little skewness is evident in the columns. But most of the are too small to be significantly evident on the plots. Therefore, we will use the .skew() to find and correct skewness later in the section. This skewness might be eliminated when we treat any outliers, if present.

```python
plt.figure(figsize=(20,20))

for index in range(len(numericCols)):
    plt.subplot(4,4,index+1)
    plt.boxplot(baseballDf[numericCols[index]], whis=3)
    plt.xlabel(numericCols[index])
    plt.tight_layout()
```

From the boxplots above, we see that there are no outliers in the data when we set the parameter for outlier is set to 3 times the interquartile range (IQR) below the 25th percentile value or above the 75th percentile value. Later in the analysis, if our model accuracy is not enough, we will try to set the outliers parameter to 1.5 times the interquartile range (IQR) below the 25th percentile value or above the 75th percentile value and treat them accordingly to increase accuracy.

The reason I am prefering 3 times the interquartile range (IQR) below the 25th percentile value or above the 75th percentile value as outliers is so that most of the data is not put as outliers. This is so because our data is very small and deleting or modifying any values may significantly affect the model.

###### Bivariate Analysis

```python
figure = plt.figure(figsize=(20,20))

for index in range(len(numericCols)):
    plt.subplot(4,4,index+1)
    sns.regplot(data = baseballDf, x = baseballDf[numericCols[index]], y = 'W')
    plt.tight_layout()
```

In the regression plots above, the y-axis represents our target variable 'W' and the x-axis represents each feature variable. We can see that columns RA, ER and ERA has very strong negative relation with W. Further more, the data points for these columns are very close to the regression line. On the other hand, R, 2B, SHO and SV have positive relation with W. But the data points are not as close to the regression lines. Lastly, AB, H, CG and E seems to have very weak relation with W since they are almost flat. Based on all this information, we can conclude that the columns RA, ER, ERA, R, 2B, SHO and SV seems to be very good predictors of the target variable W.

```python
plt.figure(figsize=(15,10))
sns.heatmap(baseballDf.corr(), vmax = 1, vmin = -1, annot= True)
```

From the correlation heatmap we can see that the columns ERA, ER, RA have very strong negative correlation with W. We also notice that SHO, BB, HR, 2B and R have a positive correlation with W and the rest have a weak correlation. This is consistent with out observation from the regression plots previously. However, we will now see if these correlations are statistically significant by calculating the p-values.

```python
#Making an empty list to store all columns with p-value less the 0.1
significantCols = []

#Calculating the pearson correlation cofficient and their p-values with respect to the target variable to see if they are
#statistically significant. If they are significant then we print those column's Pearson Correlation Coefficient and P-value
#and store the column names in the empty list initialised above
for index in range(len(numericCols)):
    pCoef, pVal = stats.pearsonr(baseballDf[numericCols[index]], baseballDf['W'])
    if pVal < 0.1:
        significantCols.append(numericCols[index])
        print(f'{numericCols[index]}:\n\tThe Pearson Correlation Coefficient:{pCoef}\n\tP-value: {pVal}')
```

## Analysis <a name = "analysis"></a>

```python
significantCols
```

```python
#Making a copy of the original dataframe but with only the significant columns
sigBaseballDf = baseballDf[significantCols].copy()
sigBaseballDf.head()
```

Now, we have two dataframes,i.e, baseballDf and sigBaseballDf. The former contains all data and the latter contains only those columns which are statistically significant. We will fit various regression algorithms to both these dataframes and see which ones have a higher R2 score. The dataframe with the higher R2 scores will be used for final model building.

We will start with baseballDf first and then sigBaseballDF.

```python
#baseballDf
```

```python
x = baseballDf.drop('W', axis = 1) #Removing the target column from assigning it to x

#making a column transformer to perform minmax scaling
colTrans = make_column_transformer((MinMaxScaler(), numericCols), remainder='passthrough') 
```

```python
x = colTrans.fit_transform(baseballDf)
```

```python
y = baseballDf.W #Specifying the target variable
```

```python
#Finding the best random state
maxAcc = 0
maxRs = 0
for randState in range(0,200):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=randState)
    lr = LinearRegression()
    lr.fit(x_train, y_train)
    predlr = lr.predict(x_test)
    acc = lr.score(x_test, y_test)*100
    if acc>maxAcc:
        maxAcc=acc
        maxRs=randState
print(f"Best accuracy is {round(maxAcc, 2)}% on Random state {randState}")
```

```python
#Splitting the x and y into train and test data with the best random state from above
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=199)
```

```python
#Fitting various regression models to the train data and printing their perfomance metrics one-by-one
lr = LinearRegression()
rfr = RandomForestRegressor()
dtr = DecisionTreeRegressor()
svr = SVR()
knr = KNeighborsRegressor()
mlp = MLPRegressor()

models = [lr, rfr, dtr, svr, knr, mlp]

for model in models:
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    print(f'{model}\n\tAccuracy: {model.score(x_test, y_test)}\n\tMSE: {mean_squared_error(y_test, pred)}\n\tRMSE: {mean_squared_error(y_test, pred)**(1/2)}\n\tR2 Score: {r2_score(y_test, pred)}')
```

Now, we will fit the regression models with on sigBaseballDf and print their performance metrics. We will then compare these performance metrics to see which dataframe performs better and use that dataframe for the final model.

```python
 #sigBaseballDf
colTransSig = make_column_transformer((MinMaxScaler(), significantCols), remainder='passthrough')

xSig = colTransSig.fit_transform(sigBaseballDf)
y = baseballDf.W
```

```python
x_trainSig, x_testSig, y_trainSig, y_testSig = train_test_split(xSig, y, test_size=0.20, random_state=199)
```

```python
lrSig = LinearRegression()
rfrSig = RandomForestRegressor()
dtrSig = DecisionTreeRegressor()
svrSig = SVR()
knrSig = KNeighborsRegressor()
mlpSig = MLPRegressor()

modelSig = [lrSig, rfrSig, dtrSig, svrSig, knrSig, mlpSig]

for model in modelSig:
    model.fit(x_trainSig, y_trainSig)
    pred = model.predict(x_testSig)
    print(f'{model}\n\tAccuracy: {model.score(x_testSig, y_testSig)}\n\tMSE: {mean_squared_error(y_test, pred)}\n\tRMSE: {mean_squared_error(y_test, pred)**(1/2)}\n\tR2 Score: {r2_score(y_test, pred)}')
```

By comparing each scores from the same algorithm but fitted on the two different dataframe we see that the dataframe with only the significant column has performed poorly. This is true for all other algorithms. For example, in case of linear regression, the R2 scores for baseballDf and sigBaseballDF 1.0 and 0.74, respectively. Therefore, we will use the full dataset, i.e, baseballDf for further model building. Since we have already fitted all the algorithms, we can preceed to cross validation stage.

## Cross Validation <a name = "crossval"></a>

```python
lrCv = cross_val_score(lr,x,y, cv = 5, scoring = 'r2')
print(f'{lr}:\n\tCross Validation Score: {lrCv.mean()}\n\tScore difference: {lrCv.mean() - 1.0}')
```

```python
rfrCv = cross_val_score(rfr,x,y, cv = 5, scoring = 'r2')
print(f'{rfr}:\n\tCross Validation Score: {rfrCv.mean()}\n\tScore difference: {0.8474389750505731-rfrCv.mean()}')
```

```python
dtrCv = cross_val_score(dtr,x,y, cv = 5, scoring = 'r2')
print(f'{dtr}:\n\tCross Validation Score: {dtrCv.mean()}\n\tScore difference: {0.9406608226567768 - dtrCv.mean()}')
```

```python
svrCv = cross_val_score(svr,x,y, cv = 5, scoring = 'r2')
print(f'{svr}:\n\tCross Validation Score: {svrCv.mean()}\n\tScore difference: {0.029827585199612705 - svrCv.mean()}')
```

```python
knrCv = cross_val_score(knr,x,y, cv = 5, scoring = 'r2')
print(f'{knr}:\n\tCross Validation Score: {knrCv.mean()}\n\tScore difference: {0.9278219824679702 - knrCv.mean()}')
```

```python
mlpCv = cross_val_score(mlp,x,y, cv = 5, scoring = 'r2')
print(f'{mlp}:\n\tCross Validation Score: {mlpCv.mean()}\n\tScore difference: {0.9809072931211651 - mlpCv.mean()}')
```

From the above results we see that Linear Regression performs the best after cross-validation. It is the ideal model for this dataset since its R2 score and cross validation R2 scores have the least difference. However, just as an additional exercise, I will also take the second best model, KNeighborsRegressor and perform hyperparameter tuning to see how high accuracy it can achieve.

## Hyperparameter Tuning <a name = "hyperparam"></a>

```python
#Linear Regression Hyperparameters
lrParameters = {'fit_intercept': [True, False],
             'normalize': [True, False],
             'positive': [True, False]}

#KNeighbors Regressor Hyperparameters
knrParameters = {'n_neighbors': [5,10, 15, 20],
                'weights': ['uniform', 'distance'],
                'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']}
```

```python
lrGS = GridSearchCV(lr, param_grid = lrParameters, cv = 5, scoring = 'r2', n_jobs = -1)
```

```python
knrGs = GridSearchCV(knr, param_grid = knrParameters, cv = 5, scoring = 'r2', n_jobs = -1)
```

```python
lrGS.fit(x_train, y_train)
```

```python
knrGs.fit(x_train, y_train)
```

```python
knrGs.best_score_
```

```python
knrGs.best_params_
```

```python
knr = KNeighborsRegressor(algorithm = 'ball_tree', n_neighbors = 5, weights = 'distance')

knr.fit(x_train, y_train)

pred = knr.predict(x_test)

print(f'R2 Score after hyperparameter tuning: {r2_score(y_test, pred)}')
```

Therefore, we see that with hyperparameter tuning, KNeighbors can achieve accuracy as high as 96.27%. However, for the purpose of this project, our final model will be Linear Regression model.

```python
lrGS.best_params_
```

```python
lr = LinearRegression(fit_intercept = True, normalize = True, positive = True)

lr.fit(x_train, y_train)

pred = lr.predict(x_test)

print(f'R2 Score after hyperparameter tuning: {r2_score(y_test, pred)}')
```
